In Python, **keywords** and **identifiers** are fundamental building blocks. Lets understand each of them with definitions and examples.

---

##  **Keywords**

**Definition:**  
Keywords are reserved words in Python that have predefined meanings and cannot be used as variable names, function names, or any other identifiers.

**Features of Keywords:**
- They are case-sensitive (e.g., `True` and `true` are different).
- Python has a fixed set of keywords that may vary slightly with different versions of Python.
- You cannot use them for naming variables, classes, or functions.

### List of Python Keywords (for Python 3.11+):

```python
False      class      from       or
None       continue   global     pass
True       def        if         raise
and        del        import     return
as         elif       in         try
assert     else       is         while
async      except     lambda     with
await      finally    nonlocal   yield
break      for        not
```

### Examples of Keywords in Code:

```python
# Using some keywords in a Python script

if True:
    print("This is a keyword example")

for i in range(5):  # 'for' is a keyword
    print(i)

def greet():  # 'def' is used to define a function
    return "Hello, World!"  # 'return' is a keyword
```

---

##  **Identifiers**

**Definition:**  
Identifiers are the names used to identify variables, functions, classes, and objects in Python.

**Rules for Identifiers:**
1. Must start with a letter (a-z, A-Z) or an underscore (`_`).
2. Can contain letters, digits (0-9), and underscores.
3. Cannot be a keyword.
4. Case-sensitive (e.g., `Age` and `age` are different).
5. Should not contain special characters like `@`, `$`, `%`.

### Examples of Valid Identifiers:

```python
name = "Alice"
_age = 25
totalAmount = 100
```

### Examples of Invalid Identifiers:

```python
1name = "Alice"      # Starts with a digit (Invalid)
total-amount = 100   # Contains a hyphen (Invalid)
class = "Python"     # 'class' is a keyword (Invalid)
```

### Examples of Identifiers in Code:

```python
def my_function():  # 'my_function' is an identifier for the function
    studentName = "John"  # 'studentName' is an identifier for a variable
    return studentName

print(my_function())
```

## what is a variable
 
 A variable is a named reference to a value stored in memory. It allows you to store data that you can use and manipulate throughout your program. Variables can hold different types of data, such as numbers, strings, lists, dictionaries, etc.

 Defining Variables
To define a variable in Python, you simply assign a value to a name using the = operator

````python
# Numeric variables
age = 25
height = 5.9

# String variables
name = "Alice"
greeting = "Hello, world!"

# Boolean variable
is_student = True

# List variable
fruits = ["apple", "banana", "cherry"]

# Dictionary variable
person = {"name": "Bob", "age": 30, "is_student": False}

# Tuple variable
coordinates = (10, 20)

# Set variable
unique_numbers = {1, 2, 3, 4, 5,"name"}
````

## eval

### Syntax:

### eval(expression, globals=None, locals=None)

First: locals (local scope)

Then: globals (global scope)

so first always locals have most priority and globals have second priority.


````python
globals_dict = {"x": 5, "y": 15}
locals_dict = {"y": 5}
exp="x==y"
a=eval(exp,globals_dict,locals_dict)
print(a)
````

````python
globals_dict = {"x": 5, "y": 15}
locals_dict = {"y": 25}
exp="x+y"
a=eval(exp,locals_dict,globals_dict) ### locals and globals just name different based on the syntax position it will descide the global and local.
print(a)
````

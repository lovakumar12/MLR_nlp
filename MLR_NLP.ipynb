```python
import nltk
nltk.download('stopwords')

import spacy
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import string

# Load SpaCy model
nlp = spacy.load("en_core_web_sm")

# Initialize NLTK tools
nltk_stop_words = set(stopwords.words("english"))
stemmer = PorterStemmer()

def process_text(text):
    # Tokenize and preprocess with SpaCy
    doc = nlp(text)
    processed_tokens = []
    lemmas = []
    stems = []

    for token in doc:
        # Remove punctuation and stop words
        if token.text in string.punctuation or token.text.lower() in nltk_stop_words:
            continue

        # Add the clean token
        processed_tokens.append(token.text)
        # Lemmatization using SpaCy
        lemmas.append(token.lemma_)
        # Stemming using NLTK
        stems.append(stemmer.stem(token.text))

    return {
        "tokens": processed_tokens,
        "lemmas": lemmas,
        "stems": stems
    }

# Example usage
if __name__ == "__main__":
    sample_text = "This is an example sentence, # demonstrating tokenization,  $ lemmatization, and stemming!"
    result = process_text(sample_text)
    print("Tokens:", result["tokens"])
    print("Lemmas:", result["lemmas"])
    print("Stems:", result["stems"])
```

##N-Gram:

An n-gram is a sequence of n consecutive words or tokens from a text.

n=3

n=3 (Trigram), the sentence "I love programming" yields:

```python
[('I', 'love', 'programming')]
```

```python
from nltk import ngrams
from collections import Counter

# Input text
text = "I love programming my name is laxman my vuru annavaram"

# Tokenize the text
tokens = text.split()

# Function to generate n-grams
def generate_ngrams(tokens, n):
    return list(ngrams(tokens, n))

# Unigrams
unigrams = generate_ngrams(tokens, 1)
print("Unigrams:", unigrams)

# Bigrams
bigrams = generate_ngrams(tokens, 2)
print("Bigrams:", bigrams)

# Trigrams
trigrams = generate_ngrams(tokens, 3)
print("Trigrams:", trigrams)

# General n-grams (n=4)
n = 4
n_grams = generate_ngrams(tokens, n)
print(f"{n}-grams:", n_grams)

# Frequency Analysis of Bigrams
bigram_counts = Counter(bigrams)
print("Bigram Frequencies:", bigram_counts)
```

**What is One-Hot Encoding?**

One-hot encoding is a technique used to convert categorical data into a numerical format that can be fed into machine learning models. It represents each unique category as a binary vector where:


A 1 indicates the presence of a specific category.
A 0 indicates the absence of that category.
This encoding avoids assigning arbitrary numeric values to categories, which could mislead models into interpreting ordinal relationships where none exist.


**Example of One-Hot Encoding**
Example 1: Encoding Colors
Suppose we have a dataset with a single feature Color containing these values:


**Advantages:**
Ensures categorical features are treated equally by algorithms.
Prevents unintended ordinal relationships between categories.

**Disadvantages:**

#### High Dimensionality:
For features with many categories, one-hot encoding can create a large number of columns (e.g., encoding 10,000 cities).

####Sparse Representation:
Most values in the encoded vector are 0, which increases memory usage.




**Sentences:**

"I love coding"

"I love Python"

"Coding is fun"

**Steps in One-Hot Encoding**

**Step 1: Tokenization**
Break each sentence into individual words or tokens:

["I", "love", "coding", "Python", "is", "fun"]

**Step 2: Vocabulary Creation**
Create a vocabulary (unique list of words):

Vocabulary: ["I", "love", "coding", "Python", "is", "fun"]

**Step 3: One-Hot Encoding**

Represent each word in the vocabulary as a binary vector.
Each vector has the length equal to the vocabulary size, and only one position is 1 for the respective word

image.png

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# Sample sentences
sentences = ["I love coding", "I love Python", "Coding is fun"]
# Tokenize sentences into words
words = [sentence.split() for sentence in sentences]
# Flatten and find unique words (vocabulary)
vocabulary = sorted(set(word for sentence in words for word in sentence))
# Create one-hot encoder
# The 'sparse' argument has been renamed to 'sparse_output' in newer versions of scikit-learn.
encoder = OneHotEncoder(sparse_output=False)
encoder.fit(np.array(vocabulary).reshape(-1, 1))

# Encode each sentence
encoded_sentences = []
for sentence in words:
    encoded = encoder.transform(np.array(sentence).reshape(-1, 1))
    encoded_sentences.append(encoded)

# Display results
print("Vocabulary:", vocabulary)
for i, sentence in enumerate(sentences):
    print(f"\nSentence: '{sentence}'")
    print("One-Hot Encoded:\n", encoded_sentences[i])
```

## Bag of words


**What is Bag of Words (BoW)?**

The Bag of Words (BoW) model is a simple and commonly used technique in Natural Language Processing (NLP) for text representation. It transforms text into numerical features by counting the frequency of each word in a document, ignoring grammar and word order.

**Key Characteristics of BoW:**

**Word Frequency-Based:** Each document is represented as a vector of word counts or occurrences.

**Orderless:** The model disregards the order of words, treating the document as a "bag" of words.

**Sparse Representation:** For a large vocabulary, many words might not appear in a document, leading to sparse matrices.


**Steps in BoW:**

1)Tokenization: Split the text into words (tokens).

2)Vocabulary Creation: Create a list of all unique words across the documents.

3)Vectorization: Represent each document as a vector of word counts, with one dimension for each word in the vocabulary.


**Advantages of BoW:**

Simple and Intuitive: Easy to implement and understand.

Efficient: Works well for small to medium-sized datasets.
Foundation for Advanced Techniques: Acts as a baseline for more complex models like TF-IDF and embeddings.

**Disadvantages of BoW:**

Ignores Context: Word order and semantics are lost.
E.g., "not good" and "good" are treated similarly.
Sparse Representation: Results in large, sparse matrices for extensive vocabularies.

Vocabulary Sensitivity: Performance can degrade if the vocabulary is too large or too small.

Dimensionality Explosion: As the number of unique words grows, so does the feature space.


Dataset:
Three example documents:

"I love programming in Python."

"Python programming is fun."

"I do not love bugs in code."

## Calculate TF-IDF Manually

```python
from collections import Counter
from math import log

# Function to compute TF
def compute_tf(text):
    words = text.split()
    word_count = Counter(words)
    total_words = len(words)
    return {word: count / total_words for word, count in word_count.items()}

# Function to compute IDF
def compute_idf(corpus):
    num_docs = len(corpus)
    all_words = set(word for text in corpus for word in text.split())
    return {
        word: log(num_docs / sum(1 for text in corpus if word in text.split()))
        for word in all_words
    }

# Function to compute TF-IDF
def compute_tfidf(tf, idf):
    return {word: tf[word] * idf[word] for word in tf}

# Preprocess corpus
preprocessed_corpus = [doc.lower().replace(".", "") for doc in corpus]

# Compute TF for each document
tf_list = [compute_tf(text) for text in preprocessed_corpus]

# Compute IDF
idf = compute_idf(preprocessed_corpus)

# Compute TF-IDF for each document
tfidf_list = [compute_tfidf(tf, idf) for tf in tf_list]

# Display results
for i, tfidf in enumerate(tfidf_list):
    print(f"Document {i+1} TF-IDF:\n{tfidf}\n")
```

## Advantages and Disadvantages of TF-IDF
## Advantages
### Simplicity and Interpretability:

TF-IDF is easy to compute and understand.
Results are interpretable since each term's weight directly reflects its importance.
### Efficient for Smaller Datasets:

Works well with small to medium-sized datasets and simple tasks like document classification and keyword extraction.
### Handles High-Frequency Terms:

Reduces the importance of common terms (e.g., "the," "is") by using IDF, ensuring more relevant terms are highlighted.
### Preprocessing Flexibility:

Can be customized by adjusting the formula or preprocessing steps (e.g., removing stop words, applying stemming).
Widely Used:

A standard approach in text mining and information retrieval systems like search engines.


##Disadvantages
### Sparsity in High Dimensions:

Creates sparse vectors, especially for large vocabularies, leading to high memory usage and computational overhead.
### Ignores Context and Semantics:

TF-IDF considers individual terms independently, ignoring the context and relationships between words.
Cannot capture phrases or word meanings (e.g., "New York" as a single entity).
### Sensitive to Vocabulary Changes:

Adding or removing documents in the corpus can significantly alter the IDF scores.
###Not Suitable for Large Corpora:

Struggles with scalability as the corpus size increases due to high dimensionality.
##Limited to Bag-of-Words Representation:

Relies on bag-of-words, ignoring word order and syntactic structure, which limits its effectiveness for advanced NLP tasks.
Performance on Long Documents:

Long documents tend to have diluted TF-IDF scores, making important terms less prominent.
###No Learning Ability:

TF-IDF is not a machine learning model and cannot improve its representation based on training data.
When to Use TF-IDF

##Good Fit:
 Text classification, keyword extraction, and search relevance for small to medium datasets.
##Limitations for:
Tasks requiring semantic understanding (e.g., sentiment analysis, machine translation). For these, embeddings like Word2Vec, GloVe, or contextual models like BERT are better.

##Word2Vec

##What is Word2Vec?
Word2Vec is a technique to learn word embeddings. Each word is mapped to a vector of real numbers in a multi-dimensional space, capturing semantic meaning. Words with similar meanings are close in this vector space.

There are two main training models:

CBOW (Continuous Bag of Words): Predicts the target word using the context words.
Skip-Gram: Predicts the context words given the target word.


```python
!pip install gensim
```

```python
# Example corpus
corpus = [
    "The quick brown fox jumps over the lazy dog",
    "A journey of a thousand miles begins with a single step",
    "To be or not to be that is the question",
    "All that glitters is not gold",
    "The early bird catches the worm"
]

# Tokenize the sentences
tokenized_corpus = [sentence.lower().split() for sentence in corpus]
print("Tokenized Corpus:")
print(tokenized_corpus)
```

```python
## train the worde2vec model
from gensim.models import word2vec
model=word2vec.Word2Vec(tokenized_corpus,vector_size=50,window=5,min_count=1,sg=0,epochs=10)
model.save("word2vec.model")
```

###sentences: Tokenized corpus.

###vector_size: Dimensionality of the word vectors.

####window: Context window size.

###min_count: Ignores words with frequency less than this.

###sg=0: Use CBOW; set sg=1 for Skip-Gram.

###epochs: Number of training iterations.

```python
## inspect the words
word_vectors=model.wv['quick']
print('Word Vector for "quick":')
print(word_vectors)
```

```python
## similar words
# Most similar words to 'quick'
similar_words = model.wv.most_similar('quick', topn=5)
print("\nMost Similar Words to 'quick':")
for word, similarity in similar_words:
    print(f"{word}: {similarity}")
```

###evaluate the model

```python

# Similarity between two words
similarity = model.wv.similarity('quick', 'lazy')
print("\nSimilarity between 'quick' and 'lazy':", similarity)
```
